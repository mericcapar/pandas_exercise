```python
import pandas as pd
import numpy as np
```

```python
s1 = pd.Series(np.random.randint(1, high=5, size=100))
s2 = pd.Series(np.random.randint(1, high=4, size=100))
s3 = pd.Series(np.random.randint(10000, high=30001, size=100))

print(s1, s2, s3)
```

```python
housemkt = pd.concat([s1,s2,s3], axis = 1 , ignore_index= True)
```

```python
housemkt
```

```python
housemkt.rename(columns = {0 : 'bedrs' , 1: 'bathrs' , 2: 'price_sqr_meter'})
```
