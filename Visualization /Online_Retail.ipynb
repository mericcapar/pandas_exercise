```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Online_Retail/Online_Retail.csv'
online_rt = pd.read_csv(url , encoding = 'latin1')
online_rt = online_rt.head(200000)
```

```python
online_rt.Country.value_counts()
```

```python
online_rt.info()
```

```python
countries = online_rt.groupby("Country").sum()
```

```python
len(countries)
```

```python
countries.sort_values(by = 'Quantity' , ascending = False)
countries
```

```python
countries = countries[1:11]
countries
```

```python
countries = countries.groupby('Country').agg({'Quantity' : 'sum'})
countries
```

```python
countries = countries.sort_values('Quantity' , ascending =False)
```

```python
countries.plot(kind='bar')

plt.xlabel('Countries')
plt.ylabel('Quantity')
plt.title('10 Countries with most orders')

plt.show()
```

```python
online_rt = online_rt[online_rt.Quantity >= 0]
```

```python
online_rt.loc[: , 'total_price'] = online_rt.Quantity * online_rt.UnitPrice
```

```python

online_rt.plot(x = 'CustomerID' ,  y = 'total_price' , kind = 'scatter')
```
