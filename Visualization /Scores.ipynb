```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data =       {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73], 
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}

df = pd.DataFrame(data, columns = ['first_name', 'last_name', 'age', 'female', 'preTestScore', 'postTestScore'])

df
```

```python
plt.scatter(df.preTestScore , df.postTestScore , s=df.age)

plt.title('preTestScore X postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```

```python
plt.scatter(df.preTestScore , df.postTestScore , s=4 * df.postTestScore , c = df.female )

plt.title('2 - preTestScore X postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```

```python
gender_counts = df['female'].value_counts()

plt.bar(['Male', 'Female'], gender_counts)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```
